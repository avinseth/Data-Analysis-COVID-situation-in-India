```python
#import some required libraries
import pandas as pd 
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("../input/covid19-in-india/covid_19_india.csv",parse_dates=["Date"],dayfirst=True)
```

```python
# take required columns
df=df[["Date","State/UnionTerritory","Cured","Deaths","Confirmed"]]
```

```python
df.head()
```

```python
today=df[df.Date=="2020/07/7"]
today
```

```python
#sorting data with respect to confirmed cases
max_confirmed_cases=today.sort_values(by="Confirmed",ascending=False)
max_confirmed_cases
```

```python
# Extract top five states with comfirmed
top_states_confirmed=max_confirmed_cases[0:5]
top_states_confirmed
```

```python
sns.set(rc={"figure.figsize":[15,8]})
sns.barplot(x="State/UnionTerritory",y="Confirmed",data=top_states_confirmed,hue="State/UnionTerritory")
plt.show()
```

```python
max_deaths_cases=today.sort_values(by="Deaths",ascending=False)
max_deaths_cases
```

```python
top_states_deaths=max_deaths_cases[0:5]
top_states_deaths
```

```python
sns.set(rc={"figure.figsize":[15,8]})
sns.barplot(x="State/UnionTerritory",y="Deaths",data=top_states_deaths,hue="State/UnionTerritory")
plt.show()
```

```python
max_cured_cases=today.sort_values(by="Cured",ascending=False)
max_cured_cases
```

```python
top_states_cured=max_cured_cases[0:5]
top_states_cured
```

```python
sns.set(rc={"figure.figsize":[15,8]})
sns.barplot(x="State/UnionTerritory",y="Cured",data=top_states_cured,hue="State/UnionTerritory")
plt.show()
```

```python
# Renaming columns names
df.columns = ["date","state","cured","deaths","confirmed"]
df
```

```python
#Maharashtra
maharashtra=df[df.state=="Maharashtra"]
maharashtra
```

```python
# visualizing death case in maharashtra
sns.set(rc={"figure.figsize":(15,8)})
sns.lineplot(x="date",y="deaths",data=maharashtra,color="red")
plt.show()
```

```python
#kerala
kerala=df[df.state=="Kerala"]
kerala
```

```python
# visualizing confirmed cases in
sns.set(rc={"figure.figsize":(20,8)})
sns.lineplot(x="date",y="confirmed",data=kerala,color="green")
plt.show()
```

```python
Tamil_Nadu=df[df.state=="Tamil Nadu"]
Delhi=df[df.state=="Delhi"]
Gujarat=df[df.state=="Gujarat"]
Uttar_Pradesh=df[df.state=="Uttar Pradesh"]
```

```python
#visualizing deaths cases in top infected states
sns.set(rc={"figure.figsize":(20,8)})
sns.lineplot(x="date",y="deaths",data=maharashtra,color="red",linewidth=8)
sns.lineplot(x="date",y="deaths",data=Tamil_Nadu,color="brown",linewidth=8)
sns.lineplot(x="date",y="deaths",data=Delhi,color="blue",linewidth=8)
sns.lineplot(x="date",y="deaths",data=Gujarat,color="green",linewidth=8)
sns.lineplot(x="date",y="deaths",data=Uttar_Pradesh,color="black",linewidth=8)
plt.legend(["maharashtra","Tamil_Nadu","Delhi","Gujarat","Uttar_Pradesh"],title="States")
plt.title("Deaths_chart",size=50)
plt.grid(color="black",linestyle="--")


plt.show()
```

```python
# visualize top states by cured from covid_19
sns.set(rc={"figure.figsize":(20,8)})
sns.lineplot(x="date",y="cured",data=maharashtra,color="red",linewidth=7)
sns.lineplot(x="date",y="cured",data=Tamil_Nadu,color="brown",linewidth=7)
sns.lineplot(x="date",y="cured",data=Delhi,color="blue",linewidth=7)
sns.lineplot(x="date",y="cured",data=Gujarat,color="green",linewidth=7)
sns.lineplot(x="date",y="cured",data=Uttar_Pradesh,color="black",linewidth=7)
plt.legend(["maharashtra","Tamil_Nadu","Delhi","Gujarat","Uttar_Pradesh"],title="States")
plt.title("Recovery_Graph",size=50)
plt.grid(color="black")


plt.show()

```
